```python
data1[" S A Production (Ton) "]=data1[" S A Production (Ton) "].astype(str)
data1["Total Production (Ton)"]=data1["Total Production (Ton)"].astype(str)

data1[" S A Production (Ton) "]=data1[" S A Production (Ton) "].str.replace(",","").astype(float)
data1["S B Production (Ton)"]=data1["S B Production (Ton)"].astype(float)
data1["Total Production (Ton)"]=data1["Total Production (Ton)"].str.replace(",","").astype(float)
data1
```

```python
top_maize_production_districts=data1.sort_values(by='Total Production (Ton)',ascending=False)
top_maize_production_districts
```

```python
import matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))

plt.subplot(1, 3, 1)
sns.histplot(data1[' S A Production (Ton) '], kde=True)
plt.title(' S A Production (Ton) ')

plt.subplot(1, 3, 2)
sns.histplot(data1['S B Production (Ton)'], kde=True)
plt.title('S B Production (Ton)')

plt.subplot(1, 3, 3)
sns.histplot(data1['Total Production (Ton)'], kde=True)
plt.title('Total Production (Ton)')

plt.tight_layout()
plt.show()

```

```python
plt.figure(figsize=(14, 7))

plt.subplot(1, 3, 1)
sns.boxplot(y=data1[' S A Production (Ton) '])
plt.title(' S A Production (Ton) ')

plt.subplot(1, 3, 2)
sns.boxplot(y=data1['S B Production (Ton)'])
plt.title('S B Production (Ton)')

plt.subplot(1, 3, 3)
sns.boxplot(y=data1['Total Production (Ton)'])
plt.title('Total Production (Ton)')

plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

rwanda_map = gpd.read_file('rwa_district (1)')
rwanda_maize = rwanda_map.merge(data1)


fig, ax = plt.subplots(1, 1, figsize=(10, 8))
rwanda_maize.plot(column='Total Production (Ton)', ax=ax, legend=True,
                  legend_kwds={'label': "Maize Production (tons)",
                               'orientation': "horizontal"},
                  cmap='YlGn')
for idx, row in rwanda_maize.iterrows():
    plt.annotate(text=row['District'],  # Replace 'district_name' with the actual column name for district names
                 xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')

# Set title and show the plot
ax.set_title('Maize Production in Rwanda by District')
plt.show()
```

```python
data1= pd.DataFrame(data1, columns=['Total Production (Ton)'])
# Calculate quartiles
q1 = data1['Total Production (Ton)'].quantile(0.25)  # First quartile (Q1)
q2 = data1['Total Production (Ton)'].quantile(0.50)  # Second quartile (Q2) or median
q3 = data1['Total Production (Ton)'].quantile(0.75)  # Third quartile (Q3)

print(f"Q1: {q1}")
print(f"Q2 (Median): {q2}")
print(f"Q3: {q3}")
```
