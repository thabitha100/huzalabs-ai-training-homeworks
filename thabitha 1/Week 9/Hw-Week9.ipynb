```python
#Analyze the number of missing values and their distribution across columns
missing_counts= data.isnull().sum()
missing_percentage = (missing_counts / len(data)) * 100

missing_data = pd.DataFrame({
    'Missing Values': missing_counts,
    'Percentage': missing_percentage
})

missing_data = missing_data[missing_data['Missing Values'] > 0]

# Display the summary of my missing data
pd.DataFrame(missing_data)
```

```python
# the distribution of the missing values 
plt.figure(figsize=(10,6))
missing_data["Missing Values"].plot(kind="bar")
plt.title("number of missing value")
plt.ylabel("number of missing value per column")
plt.xlabel("Columns with missing")
plt.show()
```

```python
# droping the rating and duration by replacing 
data3= data.dropna(subset=['rating', 'duration'])
data3
```

```python
data3['date_added'] = data3['date_added'].fillna("1-Jan-21")
data3
```

```python
df=data3.fillna("unknown")
df
```

```python
# a histogram or density plot to visualize the distribution of the "release_year" column using Matplotlib.
plt.hist(df["release_year"], bins=10, density=True, alpha=0.6, color='b')
plt.title("release year")
```

```python
#the density plot of the release_year
sns.kdeplot(df["release_year"], shade=True)
plt.title("Density Plot of Release Year")
plt.xlabel("Release Year")
plt.ylabel("Density")
plt.show()
```

```python
data3["rating"].unique()
```

```python
# Define a mapping for the ratings

rating_mapping = {
    'G': 1, 'PG': 2, 'PG-13': 3, 'R': 4, 'NC-17': 5,
    'TV-Y': 1, 'TV-Y7': 2, 'TV-G': 3, 'TV-PG': 4, 'TV-14': 5, 'TV-MA': 6, 'A': 7
}
 
# Apply the mapping to convert ratings to numerical values
df['rating_value'] = df['rating'].map(rating_mapping)

# Calculate the average rating for Movies and TV Shows
average_ratings = df.groupby('type')['rating_value'].mean()

pd.DataFrame(average_ratings)
```

```python
#Create a bar chart to compare the average ratings between these categories.

average_ratings.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Average Ratings of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.ylim(0, max(average_ratings) + 1)
plt.grid(axis='y')
plt.show()
```

```python
# Group data by type and rating, then count occurrences
rating_counts = data3.groupby(['type', 'rating']).size().unstack()
rating_counts
```

```python
# Create separate bar charts for each type (Movie/TV Show)
types = rating_counts.columns

for type in types:
  plt.bar(rating_counts.index, rating_counts[type])
  plt.xlabel('Rating')
  plt.ylabel('Count')
  plt.title(f'Rating Distribution for {type}')
  plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
  plt.show()
```

```python
# Create scatter plot with colors to distinguish categories
plt.figure(figsize=(10, 6))
colors = ['b' if cat == 'Movie' else 'g' for cat in data3['type']]
plt.scatter(data3['duration'], data3['rating'], c=colors, alpha=0.7)

# Add labels and title
plt.xlabel('duration')
plt.ylabel('rating')
plt.title('Relationship Between Duration and Rating (Movie vs. TV Show)')

# Add legend
plt.legend(['Movie', 'TV Show'])

# Display plot
plt.grid(True)
plt.show()
```
